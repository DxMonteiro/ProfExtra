# Step 0. Load Packages/Libraries

```python
import csv
import random
import json
import numpy as np
from Tools.languager import Languager
from Tools.sentimenter import Sentimenter
```

```python
input = '/home/dxmonteiro/Desktop/WORKSPACE/ProfExtra/scripts/data/Camilo_AP_frases.csv'

nome = 0
texto = 1

rows = []
fields = ['nome', 'texto', '#palavras', '#unicas', '#pontos_interrogacao', '#pontos_exclamacao', '#nomes', '#adjectivos', '#adverbios', '#verbos', 'diversidade_lexical', 'densidade_lexical', 'sentimento']'] 

languager = Languager('pt_core_news_lg')
sentimenter = Sentimenter('turing-usp/FinBertPTBR')
```

```python
with open(input, 'r') as file:
  csvreader = csv.reader(file)
  next(csvreader)
  for row in csvreader:
    nplist = np.array(row)
    
    word_analysis = languager.word_analysis(str(nplist[texto]))
    
    paper = [nplist[nome],nplist[texto],languager.num_words(str(nplist[texto])),languager.unique_words(str(nplist[texto])),languager.points(str(nplist[texto])).get('interrogation'),languager.points(str(nplist[texto])).get('exclamation'),word_analysis.get('#noun'),word_analysis.get('#adjective'),word_analysis.get('#adverb'),word_analysis.get('#verb'),word_analysis.get('lexical_diversity'),word_analysis.get('lexical_density'),sentimenter.sentiment_analysis(str(nplist[texto]))]
    rows.append(paper)
```

```python
with open('/home/dxmonteiro/Desktop/WORKSPACE/ProfExtra/scripts/data/output_camilo.csv', 'w') as csvfile:
    csvwriter = csv.writer(csvfile) 
    csvwriter.writerow(fields) 
    csvwriter.writerows(rows)
```
